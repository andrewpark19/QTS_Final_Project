```python
crypto_universe = filtered_df.select('symbol_id').unique()['symbol_id']
```

```python
start = datetime.date(2016, 1, 1).isoformat()
end = datetime.date(2025, 2, 25).isoformat()
interval = '1DAY'
all_data = {}
for ticker in crypto_universe:
    ohlcv_historical = api.ohlcv_historical_data(ticker, {'period_id': interval, 'time_start': start, 'time_end': end, 'limit': 10000})
    all_data[ticker] = ohlcv_historical
```

```python
# Convert each dictionary in all_data to a Polars DataFrame
# dataframes = [pl.DataFrame(data) for data in all_data]
pl_dfs = []

for key in all_data:
    pl_df = pl.DataFrame(all_data[key])
    pl_df = pl_df.with_columns(
        pl.lit(key).alias('symbol_id'),
        (pl.col('price_close') * pl.col('volume_traded')).alias('calculated_vol_usd')
    )
    pl_dfs.append(pl_df)
res = pl.concat(pl_dfs)
res
```

```python
min_date = res['time_period_end'].min()
min_date = datetime.datetime.strptime(min_date, '%Y-%m-%dT%H:%M:%S.%f0Z').date()
min_date_str = min_date.isoformat()
min_date_str

max_date = res['time_period_end'].max()
max_date = datetime.datetime.strptime(max_date, '%Y-%m-%dT%H:%M:%S.%f0Z').date()
max_date = max_date.isoformat()
max_date

interval = '1DAY'


res.write_parquet(f'../data/raw/raw_crypto_data_{min_date}_to_{max_date}_{interval}.parquet')
```
